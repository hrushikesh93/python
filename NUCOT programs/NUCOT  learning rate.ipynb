## Change learning rate to get the 			       proper graph/minimum point 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def g(x):
    return x**4-4*x**2+5

def gf(x):
    return 4*x**3-8*x
```

```python
x2 = np.linspace(-2.5,2.5,1000)
```

```python
def gradient_descent(derivative_func , initial_guess, multiplier, max_iter):
    new_x = initial_guess
    gamma = multiplier
    
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for i in range(max_iter):
        previous_x = new_x
        
        slope = derivative_func(previous_x)
        
        new_x = previous_x - gamma*slope

        x_list.append(new_x)
        
        slope_list.append(derivative_func(new_x))

        step_size = abs(new_x - previous_x)

    return new_x,x_list,slope_list
```

```python
local_min, list_X , deriv_list = gradient_descent(gf ,3,0.01,200)

plt.plot(x2, gf(x2) , color="red")
plt.scatter(list_X ,deriv_list,s=100, color = "green" , alpha=0.2)
```
