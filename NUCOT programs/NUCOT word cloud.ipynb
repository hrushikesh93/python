```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words=set(stopwords.words('english'))
print(stop_words)
```

```python
data="Artificial intelligence (AI) is intelligence—perceiving, synthesizing, and inferring information—demonstrated by machines, as opposed to intelligence displayed by non-human animals and humans. Example tasks in which this is done include speech recognition, computer vision, translation between (natural) languages, as well as other mappings of inputs."
```

```python
data
```

```python
def data_preprocessing(data):
    #lowercase
    data=data.lower()
    #tokenize the data
    data_tokens=word_tokenize(data)
    #remove stopwords
    processed_words=[w for w in data_tokens if not w in stop_words]
    return " ".join(processed_words)
```

```python
#apply processing to data
data_preprocessed=data_preprocessing(data)
```

```python
data
```

```python
data_preprocessed
```

```python
#ploting a wordcloud
plt.figure(figsize=(10,10),facecolor='None')
wordcloud=WordCloud().generate(data_preprocessed)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
#saving wordcloud as png
import os
os.chdir('F:\wc')
wordcloud.to_file("wc1.png")
```

```python
#word cloud shape
from PIL import Image
```

```python
mask=np.array(Image.open('brain.jpg'))
```

```python
maskwordcloud=WordCloud(background_color='white',mask=mask)
maskwordcloud.generate(data_preprocessed)
plt.figure(figsize=(6,6))
plt.imshow(maskwordcloud)
plt.axis('off')
plt.show()
```

```python
quotes_data=pd.read_excel('quotes.xlsx')
print(quotes_data)
```

```python
quotes_data.quotes=quotes_data['quotes'].apply(data_preprocessing)
```

```python
quote_text=' '.join([w for w in quotes_data['quotes']])
plt.figure(figsize=(10,10),facecolor='None')
wordcloud=WordCloud().generate(quote_text)
plt.axis('off')
plt.imshow(wordcloud)
plt.show()
```

```python
#apply mask 
mwc=WordCloud(background_color='white',mask=mask)
mwc.generate(quote_text)
plt.figure(figsize=(10,10))
plt.imshow(mwc)
plt.axis('off')
plt.show()
```

```python
mwc.to_file('whitewordcloud.jpeg')
```

```python
#transperant background color for mask wordcloud
mwct=WordCloud(background_color=None,mask=mask,mode='RGBA')
mwct.generate(quote_text)
plt.figure(figsize=(10,10))
plt.imshow(mwct)
plt.axis('off')
plt.show()
```

```python
mwct.to_file('whitewordcloudtrans.png')
```

```python
#outline for masks
mwct=WordCloud(background_color='white',mask=mask,contour_width=0.5,contour_color='red')
mwct.generate(quote_text)
plt.figure(figsize=(10,10))
plt.imshow(mwct)
plt.axis('off')
plt.show()
```
