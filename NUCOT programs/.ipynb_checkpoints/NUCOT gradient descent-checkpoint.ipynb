## gradient descent

```python
new_x = 3
gamma = 0.1 #learming rate

for i in range(100):
    previous_x = new_x
    slope = df(previous_x)
    new_x = previous_x - gamma*slope
    
print(round(new_x,2))
```

```python
new_x = 3
gamma = 0.1

x_list = [new_x]
slope_list = [df(new_x)]

for i in range(100):
    previous_x = new_x
    slope = df(previous_x)
    new_x = previous_x - gamma*slope
    
    x_list.append(new_x)
    
    slope_list.append(df(new_x))
    
print("local minimum: ",round(new_x,2))
```

```python
#gradient descent as a python function
```

```python
def gradient_descent(derivative_func , initial_guess, multiplier , precesion , max_iter):
    new_x = initial_guess
    gamma = multiplier
    precision = precesion
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for i in range(max_iter):
        previous_x = new_x
        
        slope = derivative_func(previous_x)
        
        new_x = previous_x - gamma*slope

        x_list.append(new_x)
        
        slope_list.append(derivative_func(new_x))

        step_size = abs(new_x - previous_x)

        if step_size < precision:
            print("the loop ran these many times:....",i,'\n')
            break

    return new_x,x_list,slope_list


local_min, list_X , deriv_list = gradient_descent(df ,3,0.1,0.001,200)
```

```python
#======================================================
# my gradient descent ( "i tired this" )
#======================================================


def f(x):
    return x**5-2*x**4+2

def df(x):
    return 5*x**4-8*x**3

x2 = np.linspace(-2.5,2.5,1000)

plt.figure(figsize=(18,8))

plt.subplot(2,2,1)
plt.title("x^5-2x^4+2")
plt.plot(x2,f(x2))
plt.xlim(-1,2)      # to zoom at particular limit in x direction
plt.ylim(-4,4)      # to zoom at particular limit in y direction
plt.grid()

plt.subplot(2,2,2)
plt.title("5x^4-8x^3")
plt.plot(x2,df(x2), color="pink")
plt.xlim(-1,2)
plt.ylim(-4,4)
plt.grid()
#----------------------------------------------------------------------
new_x = 1.9
gamma = 0.01
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for i in range(100):
    previous_x = new_x
    slope = df(previous_x)
    new_x = previous_x - gamma*slope
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    step_size = abs(new_x - previous_x)
    
    if step_size < precision:
        print("the loop ran these many times:....",i,'\n')
        break
    
print("local minimum: ",round(new_x,4))
#----------------------------------------------------------------------

plt.subplot(2,2,3)
plt.plot(x2, f(x2))
values = np.array(x_list)
plt.scatter(x_list,f(values), color = "black" , alpha=0.6)
plt.xlim(-1,2)
plt.ylim(-4,4)
plt.grid()

plt.subplot(2,2,4)
plt.plot(x2, df(x2) , color="pink")
plt.scatter(x_list ,slope_list,s=100, color = "black" , alpha=0.6)
plt.xlim(-1,2)
plt.ylim(-4,4)
plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from math import log
```

```python
"""
def f(x,y):
    r = 3**(-x**2 -y**2)
    return 1/(r+1)
"""

def f(x,y):
    
    r = 3**(-x**2 - y**2)
    
    return 1/(r+1)
```

```python
x_4 = np.linspace(start=-2, stop=2, num=200)
y_4 = np.linspace(start=-2, stop=2, num=200)

x_4.shape

y_4.shape

x_4, y_4 = np.meshgrid(x_4, y_4)

x_4.shape

y_4.shape

fig=plt.figure(figsize = (7,7)).add_subplot(projection = '3d')



fig.set_xlabel('X', fontsize = 20, c = 'red')

fig.set_ylabel('Y', fontsize = 20, c = 'red')

fig.set_zlabel('f(X,Y)....Cost Function', fontsize = 20, c = 'red')

fig.plot_surface(x_4, y_4, f(x_4, y_4), alpha =0.3, cmap = cm.coolwarm)



plt.show()
```
