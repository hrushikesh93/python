```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words=set(stopwords.words('english'))
print(stop_words)
```

```python
example = 'spam.txt'

stream = open(example, encoding='latin-1')
is_body = False
lines = []

for line in stream:
    if is_body:
        lines.append(line)
    elif line == '\n':
        is_body = True

stream.close()
```

```python
lines
```

```python
email_body = ''.join(lines)
```

```python
print(email_body)
```

```python
def data_preprocessing(data):
    #lowercase
    data=data.lower()
    #tokenize the data
    data_tokens=word_tokenize(data)
    #remove stopwords
    processed_words=[w for w in data_tokens if not w in stop_words]
    return " ".join(processed_words)
```

```python
data_preprocessed=data_preprocessing(email_body)
```

```python
data_preprocessed
```

```python
plt.figure(figsize=(10,10),facecolor='None')
wordcloud=WordCloud().generate(data_preprocessed)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
from PIL import Image

mask=np.array(Image.open('hands.jpg'))

maskwordcloud=WordCloud(background_color='white',mask=mask,contour_color='green',contour_width=0.2)
maskwordcloud.generate(data_preprocessed)
plt.figure(figsize=(16,16))
plt.imshow(maskwordcloud)
plt.axis('off')
plt.show()
```

```python
maskwordcloud.to_file('whitewordcloud.jpeg')
```

```python
from PIL import Image

mask=np.array(Image.open('apple.jpg'))

maskwordcloud=WordCloud(background_color='white',mask=mask,contour_color='green',contour_width=0.2)
maskwordcloud.generate(data_preprocessed)
plt.figure(figsize=(16,16))
plt.imshow(maskwordcloud)
plt.axis('off')
plt.show()

maskwordcloud.to_file('whitewordcloud2.jpeg')
```
