### simple linear regression

```python
# imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
import warnings
warnings.simplefilter('ignore')
```

```python
# getting data
data = pd.read_csv('heart_disease_dataset.csv')
data
```

```python
age = data[['age']]
chol = data[['chol']]
age
```

```python
age.ndim
```

```python
age.shape
```

```python
regx = LinearRegression()
regx.fit(age,chol)
```

```python
regx.intercept_[0] #intercept
```

```python
regx.coef_[0][0]
```

```python
age.values
```

```python
y_slope = regx.coef_[0][0]*age.values + regx.intercept_[0] # slope
y_slope
```

```python
y_predicted = regx.predict(age)
y_predicted
```

```python
# plot the line of regression...along with the scatter points..


plt.figure(figsize = (14,5))
plt.title('SCATTER and REGRESSION OF AGE vs CHOL', fontsize = 22)
plt.xlabel('age',fontsize = 20 )
plt.ylabel('chol', fontsize = 20)
plt.text(x=5.1, y = 74, s = 'plot', fontsize = 20)
plt.scatter(age,chol, color = 'green', s=100, alpha = 0.9)

plt.plot(age, regx.predict(age), color = 'black')
plt.show()
```

```python
regx.score(age,chol) #score
```

```python
plt.scatter(chol,y_predicted)
plt.plot()
```

```python
# linear regression ( class work )

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
import warnings
warnings.simplefilter('ignore')


data = pd.read_csv('cost_revenue_clean.csv')
data
```

```python
x = data[['production_budget_usd']]
x
```

```python
y = data[['worldwide_gross_usd']]
y
```

```python
regx2 = LinearRegression()
regx2.fit(x,y)

coef = regx2.coef_[0][0]
intercept = regx2.intercept_[0]
values = x.values

slope = coef*values + intercept
score = regx2.score(x,y)

print("[[coeficient]] \n",coef)
print("\n [[intercept]] \n",intercept)
print("\n [[values]] \n",values)
print("\n [[slope]] \n",slope)
print("\n [[score]] \n",score)

plt.figure(figsize=(20,10))
plt.scatter(x,y,color = 'blue')
plt.plot(x,slope,color = 'red')
plt.title("pbu v/s wgu")
plt.xlabel("production budget usd")
plt.ylabel("worldwide gross usd")
```
